### Code optimizations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
optimizations = [
    "Baseline", "TF32", "Compile", "Combined-qkv-proj", "Flash Attention",
    "Without Flash Attn", "bfloat16 (same batch size)", "Flash Attn + bfloat16",
    "Increase Batch Size (=16)", "Gradient Accumulation"
]
tokens_per_sec = [
    17048.7, 20597.4, 27216.3, 28582.9, 28044.8,
    28834.2, 49004, 61324.4, 67670.9, 69163.8
]

# Plot
plt.figure(figsize=(14, 7))
plt.barh(optimizations, tokens_per_sec, color='skyblue')
plt.xlabel('Tokens/s')
plt.title('Code Optimization Techniques and Their Tokens/s Performance')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

### Plot Learning Rate Scheduler

```python
import numpy as np
import matplotlib.pyplot as plt

from src.utils import CosineLRScheduler
```

```python

lr_scheduler = CosineLRScheduler(
    optimizer=None,
    max_lr=0.1,
    min_lr=0.01,
    warmup_steps=100,
    total_steps=1000
)


steps = np.arange(0, 1000)
lr_vals = [lr_scheduler.get_lr(x) for x in steps]

fig = plt.figure(figsize=(10, 5))
plt.plot(steps, lr_vals, lw=3)
plt.xlabel('Steps', fontsize=15)
plt.ylabel('Learning Rate', fontsize=15)
plt.show()
```
